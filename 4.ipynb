### Рост взрослого населения города X имеет нормальное распределение. Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см. Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:

* больше 182 см?
* больше 190 см?
* от 166 см до 190 см?
* от 166 см до 182 см?
* от 158 см до 190 см?
* не выше 150 см и не ниже 190 см?
* не выше 150 см и не ниже 198 см?
* ниже 166 см?

Решение:

Z = (x - mu) / sigma, где mu = 174, а sigma = 8, место x подставляем значения.

Далее смотрим в таблице нормального распределения (по умолчанию X <= x)

Если требуется найти "больше", то используем обратную таблицу распределения

```python
def Z(num, mu=174, sigma=8):
    return (num - mu) / sigma


p_1 = Z(182) # 0.1587 = 15.87%
p_2 = Z(190) # 0.0228 = 2.28% 
p_3 = Z(190) - Z(166) # 0.9772 - 0.1587 = 0.8185 = 81.85%
p_4 = Z(182) - Z(166) # 0.8413 - 0.1587 = 0.6826 = 68.26%
p_5 = Z(190) - Z(158) # 0.9772 - 0.0228 = 0.9544 = 95.44%
p_6 = Z(150) + Z(190) # 0.0013 + 0.0228 = 0.0241 = 2.41%
p_7 = Z(150) + Z(198) # 0.0013 + 0.0013 = 0.0026 = 0.26%
p_8 = Z(166) # 0.1587 = 15.87%
```

Ответ: Вероятность, того, что случайным образом выбранный взрослый человеĸ имеет рост:

* больше 182 см = 15.87%
* больше 190 см = 2.28%
* от 166 см до 190 см = 81.85%
* от 166 см до 182 см = 68.26%
* от 158 см до 190 см = 95.44%
* не выше 150 см и не ниже 190 см = 2.41%
* не выше 150 см и не ниже 198 см = 0.26%
* ниже 166 см = 15.87%
