### Непрерывная случайная величина X распределена нормально и задана плотностью распределения f(x) = (1 / (4 * sqrt(2pi))) * exp((-(x+2)**2) / 32) Найдите: а). M(X) б). D(X) в). std(X) (среднее квадратичное отклонение)

a) M(X)

В общем виде плотность нормального распределения имеет вид: $$ f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e ^ {- \frac{(x - a)^2}{2 \sigma^2}} $$

где a = M(X),

следовательно M(X) = -2


b) D(X)

$$ D(X)=\frac{32}{2}=16 $$

c) среднее квадратичное отклонение

$$ \sigma=\sqrt{D(X)}=\sqrt{16}=4 $$
